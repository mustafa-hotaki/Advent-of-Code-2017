# Advent of Code 2017 - Day 9

```
data = []
with open('inputs_day_9.txt', 'r') as f:
  for line in f:
    data.append(line.strip())

text = data[0]
print(len(text))
print(text)
```

## Part 1

```
# First, we can get ride of any "!" and the character following it
#text = '{{<!!>},{<!!>},{<!!>},{<!!>}}'
index = 0
filtered_text = ''
while(index < len(text)):
  if text[index] == '!':
    index += 2
  else:
    filtered_text += text[index]
    index += 1

print(filtered_text)
```

```
# Next we can get rid of garbage
garbage_counter = 0
garbage = False
index = 0
clean_text = ''

while(index < len(filtered_text)):
  if not garbage and filtered_text[index] == '<': # If opening and not opened before
    garbage = True

  elif garbage and filtered_text[index] == '>':
    garbage = False

  elif not garbage: # Second condition to not add '<' if the last condition was true
    clean_text += filtered_text[index]
  else:
    garbage_counter += 1

  index += 1

print(clean_text)
```

Now, that the text is cleaned up from the garbage, we can score it. My first instinct was to recursively find the score in the following fashion:

```
{{{},{},{{}}}}
=>
{              } 1
 {            } 2
  {}, {}, {  } 3 + 3 + 3
           {} 4
```

But, it is actually easier to iterate over characters and keep track of how deeply nested we are. Incrementing depth when encountering an opening brace and decrementing it when it is closed.

```
depth = 1
score = 0
for c in clean_text:
  if c == '{':
    score += depth
    depth += 1
  if c == '}':
    depth -= 1

score
```
